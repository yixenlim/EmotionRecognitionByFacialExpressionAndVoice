```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns
```

# Facial expression

```python
model_list = ['CNN','DBN','KNN','AdaBoost','Voting Classifier','SVM','SRC']
count_list = [2,1,1,1,1,1,1]
percentage_count_list = [count/sum(count_list)*100 for count in count_list]

plt.bar(model_list, percentage_count_list, align='center')
plt.xticks(rotation ='45')
plt.title('''Popularity of different models 
in emotion recognition by facial expression''')
plt.xlabel("Model types")
plt.ylabel("Popularity(%)")
plt.tight_layout()
plt.savefig('Count_chart_fe.png')
plt.show()
```
